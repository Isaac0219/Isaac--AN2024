Tomando el caso anterior para una aproximación por series de taylor, se tomará el rango y se seccionara en mas pequeños y se harán aproximaciónes alrededor de puntos centrales en cada sección de estos.

Asi que de nueva cuenta definimos el rango de la función, para poder hacer una comparación será el mismo rango que en la tarea anterior [-2,2], por lo que tomaremos las nuevas secciones espaciadas por un a unidad.

S_1=[-2,-1]

```python
import numpy as np
import sympy as sy
import matplotlib.pyplot as plt
```

Definimos los simbolos que usaremos, como haremos distintas aproximaciones las definimos:

```python
x,y,z = sy.symbols('x y z')
x = sy.symbols('x')
f = (sy.exp(x**2)-(3*sy.cos(x))**2)
g = sy.series(f,x,-1.5,5)
h = sy.series(f,x,-0.5,5)
i = sy.series(f,x,0.5,5)
j = sy.series(f,x,1.5,5)
```

Quitamos el error en notación O de todas las aproximaciones para poder transformarlas

```python
gx= g.removeO()
hx= h.removeO()
ix= i.removeO()
jx= j.removeO()
```

Transformamos nuestras funciones simbolicas a numéricas

```python
flam = sy.lambdify(x,f)
glam = sy.lambdify(x,gx)
hlam = sy.lambdify(x,hx)
ilam = sy.lambdify(x,ix)
jlam = sy.lambdify(x,jx)
```

Evaluamos nuestras aproximaciones en un punto dentro de su rango.

```python
print(flam(-1.7))
print(glam(-1.7))
```

```python
print(flam(-0.2))
print(hlam(-0.2))
```

```python
print(flam(0.9))
print(ilam(0.9))
```

```python
print(flam(1.8))
print(jlam(1.8))
```

Calculamos los errores de cada aproximación

```python
eabsg = np.abs(flam(-1.7) - glam(-1.7))
```

```python
eabsh = np.abs(flam(-0.9) - hlam(-0.9))
```

```python
eabsi = np.abs(flam(0.9) - ilam(0.9))
```

```python
eabsj = np.abs(flam(1.8) - jlam(1.8))
```

```python
gerel = np.abs((eabsg/flam(-1.7))*100)
```

```python
herel = np.abs((eabsh/flam(-0.9))*100)
```

```python
ierel = np.abs((eabsi/flam(0.9))*100)
```

```python
jerel = np.abs((eabsj/flam(1.8))*100)
```

```python
print(eabsg)
print(gerel,"%")
```

```python
print(eabsh)
print(herel,"%")
```

```python
print(eabsi)
print(ierel,"%")
```

```python
print(eabsj)
print(jerel,"%")
```

Vemos que al seccionar el rango de la aproximación por series de taylor, encontrar una aproximacion dentro de la tolerancia de 3% para valores dentro de los rango seccionados es mucho mas senncillo ya que proponiendo un orden pequeño de la serie de taylor que es 5 todas las aproximaciones entraron. y viendo los datos realmente no las únicas gráficas que es realmente nos aportarian algo serían las de comparacion de los error

Para S_1:

```python
X = np.linspace(-2,-1)
g = sy.series(f,x,-1.5,5)
gx = g.removeO()
glam = sy.lambdify(x,gx)
YF = flam(X)
YG = glam(X)
Yerr = np.abs(YF-YG)
Yord = X**5
plt.plot(X,Yerr, label = "Error", color = "Red")
plt.plot(X,Yord, label = "Orden de la función O", color = "Black")
plt.legend(loc=2)
plt.xlabel("X")
plt.ylabel("Error,G(X)")
plt.title("Gráfica del error")
plt.grid(True)
```

Para S_2

```python
X = np.linspace(-1,-0)
h = sy.series(f,x,-0.5,5)
hx = h.removeO()
hlam = sy.lambdify(x,hx)
YF = flam(X)
YH = hlam(X)
Yerr = np.abs(YF-YH)
Yord = X**5
plt.plot(X,Yerr, label = "Error", color = "Red")
plt.plot(X,Yord, label = "Orden de la función O", color = "Black")
plt.legend(loc=2)
plt.xlabel("X")
plt.ylabel("Error,G(X)")
plt.title("Gráfica del error")
plt.grid(True)
```

Para S_3

```python
X = np.linspace(0,1)
i = sy.series(f,x,0.5,5)
ix = i.removeO()
ilam = sy.lambdify(x,ix)
YF = flam(X)
YI = ilam(X)
Yerr = np.abs(YF-YI)
Yord = X**5
plt.plot(X,Yerr, label = "Error", color = "Red")
plt.plot(X,Yord, label = "Orden de la función O", color = "Black")
plt.legend(loc=2)
plt.xlabel("X")
plt.ylabel("Error,G(X)")
plt.title("Gráfica del error")
plt.grid(True)
```

Para S_4

```python
X = np.linspace(1,2)
j = sy.series(f,x,1.5,5)
jx = j.removeO()
jlam = sy.lambdify(x,jx)
YF = flam(X)
YJ= jlam(X)
Yerr = np.abs(YF-YJ)
Yord = X**5
plt.plot(X,Yerr, label = "Error", color = "Red")
plt.plot(X,Yord, label = "Orden de la función O", color = "Black")
plt.legend(loc=2)
plt.xlabel("X")
plt.ylabel("Error,G(X)")
plt.title("Gráfica del error")
plt.grid(True)
```
